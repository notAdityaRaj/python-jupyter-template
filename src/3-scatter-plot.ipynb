### Exercise 3: Scatter Plot with Annotations

```python
# TODO: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.

import matplotlib.pyplot as plt
x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

plt.scatter(x_scatter, y_scatter, color='orange', label='y = 2x')

for i, txt in enumerate(zip(x_scatter, y_scatter)):
    plt.annotate(f'{txt}', (x_scatter[i], y_scatter[i]), textcoords="offset points", xytext=(5,5), ha='center')

plt.title('Scatter Plot with Annotations')
plt.xlabel('x values')
plt.ylabel('y values')
plt.legend()

plt.show()
```
